# 3-qubit

```python
qr = QuantumRegister(3, name='q')
cr = ClassicalRegister(3, name='c')
qc = QuantumCircuit(qr)
qc.h(0)
qc.cx(0,1)
qc.cx(1,2)



%matplotlib inline
qc.draw(output='mpl')
```

```python
psi = qi.Statevector.from_instruction(qc)
psi.draw('latex', prefix='|\\psi\\rangle = ')
```

```python
qobj = assemble(qc) 
sim = Aer.get_backend('statevector_simulator')
results = sim.run(qobj).result().get_counts()
plot_histogram(results, title = 'Measurement Probabilities')
```

```python
provider = IBMQ.get_provider(hub='ibm-q-lanl', group='lanl', project='quantum-optimiza')
device = provider.get_backend('ibmq_belem') 
job = execute(qc, backend= device, shots=15000)
job_monitor(job)
print(job.job_id())

device_result = job.result()
plot_histogram(device_result.get_counts(qc), title = 'IBMQ Belem')
```

# 4-qubit

```python
qr = QuantumRegister(4, name='q')
cr = ClassicalRegister(4, name='c')
qc = QuantumCircuit(qr,cr)
qc.h(0)
qc.cx(0,1)
qc.cx(1,2)
qc.cx(2,3)
qc.barrier()
qc.measure(0,0)
qc.measure(1,1)
qc.measure(2,2)
qc.measure(3,3)

%matplotlib inline
qc.draw(output='mpl')
```

```python
qobj = assemble(qc) 
sim = Aer.get_backend('statevector_simulator')
results = sim.run(qobj).result().get_counts()
plot_histogram(results, title = 'Measurement Probabilities')
```

```python
provider = IBMQ.get_provider(hub='ibm-q-lanl', group='lanl', project='quantum-optimiza')
device = provider.get_backend('ibmq_belem') 
job = execute(qc, backend= device, shots=15000)
job_monitor(job)
print(job.job_id())

device_result = job.result()
plot_histogram(device_result.get_counts(qc), title = 'IBMQ Belem')
```

# 5-qubit

```python
qr = QuantumRegister(5, name='q')
cr = ClassicalRegister(5, name='c')
qc = QuantumCircuit(qr,cr)
qc.h(0)
qc.cx(0,1)
qc.cx(1,2)
qc.cx(2,3)
qc.cx(3,4)
qc.barrier()
qc.measure(0,0)
qc.measure(1,1)
qc.measure(2,2)
qc.measure(3,3)
qc.measure(4,4)

%matplotlib inline
qc.draw(output='mpl')
```

```python
simulator= Aer.get_backend('qasm_simulator') 
sim_result = execute(qc, backend = simulator, shots=300000).result()
plot_histogram(sim_result.get_counts(qc), title='Measurement Probabilities')
```

```python
provider = IBMQ.get_provider(hub='ibm-q-lanl', group='lanl', project='quantum-optimiza')
device = provider.get_backend('ibmq_lima') 
job = execute(qc, backend= device, shots=15000)
job_monitor(job)
print(job.job_id())

device_result = job.result()
plot_histogram(device_result.get_counts(qc), title = 'IBMQ Belem')
```

```python
from qiskit.providers.ibmq import least_busy
provider = IBMQ.get_provider(hub='ibm-q-lanl', group='lanl', project='quantum-optimiza')
backend = least_busy(provider.backends(filters=lambda b: b.configuration().n_qubits >= 5 and
                                   not b.configuration().simulator and b.status().operational==True)) 
job = execute(qc, backend= backend, shots=15000)
```

```python
device_result = job.result()
plot_histogram(device_result.get_counts(qc), title = 'IBMQ Belem')
```
